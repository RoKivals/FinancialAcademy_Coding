```python
def check_brackets(s):
    stack = []
    for c in s:
        if c == '(':
            stack.append(c)
        elif c == ')':
            if not stack or stack[-1] != '(':
                return False
            stack.pop()
    return not stack

code = '((dsasdas((das(d)das)d)asd)da)'
print(check_brackets(code))

code = '(f(gfhd((hgfdg((fghh)dads)das))ds)adddads())))'
print(check_brackets(code)) 
```

```python
class Stack:
    def __init__(self, max_size):
        self.max_size = max_size
        self.items = []

    def push(self, item):
        if len(self.items) < self.max_size:
            self.items.append(item)
        else:
            raise Exception('Stack overflow')

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            raise Exception('Stack is empty')

    def top(self):
        if not self.is_empty():
            return self.items[-1]
        else:
            raise Exception('Stack is empty')

    def is_empty(self):
        return len(self.items) == 0

    def __len__(self):
        return len(self.items)

s = Stack(5)
print(s.is_empty()) 

s.push(1)
s.push(2)
s.push(3)
print(len(s)) 

print(s.top())
print(s.pop()) 

s.push(4)
s.push(5)
print(s.top())

try:
    s.push(6) 
except Exception as e:
    print(e)

while not s.is_empty():
    print(s.pop()) 
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current_node = self.head
        while current_node.next:
            current_node = current_node.next
        current_node.next = new_node

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def delete_with_value(self, data):
        if not self.head:
            return
        if self.head.data == data:
            self.head = self.head.next
            return
        current_node = self.head
        while current_node.next:
            if current_node.next.data == data:
                current_node.next = current_node.next.next
                return
            current_node = current_node.next

    def __len__(self):
        length = 0
        current_node = self.head
        while current_node:
            length += 1
            current_node = current_node.next
        return length

    def __str__(self):
        nodes = []
        current_node = self.head
        while current_node:
            nodes.append(str(current_node.data))
            current_node = current_node.next
        return " - ".join(nodes)
linked_list = LinkedList()
linked_list.append(1)
linked_list.append(2)
linked_list.append(3)
linked_list.prepend(0)
print(linked_list)
print(len(linked_list)) 
linked_list.delete_with_value(2)
print(linked_list)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
            return
        new_node.prev = self.tail
        self.tail.next = new_node
        self.tail = new_node

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head.prev = new_node
        self.head = new_node

    def delete_with_value(self, data):
        if not self.head:
            return
        if self.head.data == data:
            self.head = self.head.next
            self.head.prev = None
            return
        if self.tail.data == data:
            self.tail = self.tail.prev
            self.tail.next = None
            return
        current_node = self.head
        while current_node:
            if current_node.data == data:
                current_node.prev.next = current_node.next
                current_node.next.prev = current_node.prev
                return
            current_node = current_node.next

    def __len__(self):
        length = 0
        current_node = self.head
        while current_node:
            length += 1
            current_node = current_node.next
        return length

    def __str__(self):
        nodes = []
        current_node = self.head
        while current_node:
            nodes.append(str(current_node.data))
            current_node = current_node.next
        return " - ".join(nodes)
        
class Queue:
    def __init__(self):
        self.items = DoublyLinkedList()

    def enqueue(self, data):
        self.items.append(data)

    def dequeue(self):
        if len(self.items) == 0:
            return None
        data = self.items.head.data
        self.items.delete_with_value(data)
        return data

    def __len__(self):
        return len(self.items)

    def __str__(self):
        return str(self.items)

queue = Queue()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
print(queue)
print(len(queue)) 
queue.dequeue()
print(queue) 
```

```python
class DynamicArray:
    def __init__(self):
        self.arr = []
        self.length = 0

    def append(self, e):
        self.arr.append(e)
        self.length += 1

    def get(self, index):
        if index < 0 or index >= self.length:
            raise IndexError("Index out of range")
        return self.arr[index]

    def set(self, index, e):
        if index < 0 or index >= self.length:
            raise IndexError("Index out of range")
        self.arr[index] = e

    def __len__(self):
        return self.length

    def __str__(self):
        return str(self.arr)
```

```python
from collections import deque

def alternate_queues(queue1, queue2):
    while True:
        try:
            yield queue1.pop()
        except IndexError:
            break
        try:
            yield queue2.pop()
        except IndexError:
            break

queue1 = deque([1, 2, 3])
queue2 = deque([4, 5, 6, 7])

for value in alternate_queues(queue1, queue2):
    print(value)
```

```python
from collections import deque

class Pump:
    def __init__(self, queue, generator):
        self.queue = queue
        self.generator = generator
        
    def action(self):
        try:
            value = next(self.generator)
            self.queue.append(value)
        except StopIteration:
            pass

class MultiAction:
    def __init__(self, n, pump_class, queue, generator):
        self.n = n
        self.pump_class = pump_class
        self.queue = queue
        self.generator = generator
        
    def action(self):
        for i in range(self.n):
            pump = self.pump_class(self.queue, self.generator)
            pump.action()

class MultiPump:
    def __init__(self, queues, generator):
        self.queues = queues
        self.generator = generator
        self.queue_index = 0
        
    def action(self):
        try:
            value = next(self.generator)
            self.queues[self.queue_index].append(value)
            self.queue_index = (self.queue_index + 1) % len(self.queues)
        except StopIteration:
            pass

queue1 = deque()
queue2 = deque()
generator = alternate_queues(deque([1, 2, 3]), deque([4, 5, 6, 7]))

pump1 = Pump(queue1, generator)
pump2 = Pump(queue2, generator)

multi_pump = MultiPump([queue1, queue2], generator)
multi_action = MultiAction(3, Pump, queue1, generator)

pump1.action()
pump2.action()
multi_pump.action()
multi_action.action()

print(queue1)
print(queue2)
```

```python
from collections import deque
import random

class GenFromQ:
    def __init__(self, queue):
        self.queue = queue
        
    def __iter__(self):
        return self
    
    def __next__(self):
        if len(self.queue) > 0:
            return self.queue.pop()
        else:
            raise StopIteration

class Pump:
    def __init__(self, queue, generator):
        self.queue = queue
        self.generator = generator
        
    def action(self):
        try:
            value = next(self.generator)
            self.queue.append(value)
        except StopIteration:
            pass

class MultiPump:
    def __init__(self, queues, generator):
        self.queues = queues
        self.generator = generator
        self.queue_index = 0
        
    def action(self):
        try:
            value = next(self.generator)
            self.queues[self.queue_index].append(value)
            self.queue_index = (self.queue_index + 1) % len(self.queues)
        except StopIteration:
            pass

queue1 = deque()
queue2 = deque()
queue3 = deque()
generator1 = GenFromQ(deque(["Message1", "Message2", "Message3"]))
generator2 = GenFromQ(deque(["Message4", "Message5", "Message6"]))
generator3 = GenFromQ(deque(["Message7", "Message8", "Message9"]))
pump1 = Pump(queue1, generator1)
pump2 = Pump(queue2, generator2)
pump3 = Pump(queue3, generator3)

multi_pump = MultiPump([queue1, queue2, queue3], random.choice([generator1, generator2, generator3]))
multi_pump.action()

merged_queue = deque()
merged_pump = Pump(merged_queue, GenFromQ(multi_pump.queues))
merged_pump.action()

print(merged_queue) 
```
